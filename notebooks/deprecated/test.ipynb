```python
# ---------------- Loss Functions ----------------
import optax
from tqdm import tqdm


def loss(model, coords, areas, target_density, lam, penalty):
    rho = jax.nn.sigmoid(model(coords))
    J = J_total(rho)
    mean_rho = (np.sum(rho.squeeze() * areas) / total_area)
    C_raw = mean_rho - target_density
    C = np.maximum(C_raw, 0.0)
    penalty_term = penalty * C**2
    lagrangian_term = lam * C
    vol_cond = lagrangian_term + penalty_term
    loss_val = J + vol_cond

    jax.debug.print(
        "mean(rho)={mean_rho:.4f} | C={C:.4f} | J={J:.4f} | vol={vol_cond:.4f} | loss={loss:.4f}",
        mean_rho=mean_rho, C=C, J=J, vol_cond=vol_cond, loss=loss_val
    )

    return loss_val, C


loss_and_grad = eqx.filter_value_and_grad(loss, has_aux=True)
loss_and_grad = eqx.filter_jit(loss_and_grad)

def optimisation_step(
    model, opt_state, optimizer, coords, areas, target_density, lam, penalty
):
    """
    One optimization step for a single model.
    """
    (loss_val, C), grads = loss_and_grad(
        model, coords, areas, target_density, lam, penalty
    )

    # If loss is not finite, just skip the update and return as-is
    if not np.isfinite(loss_val):
        return model, opt_state, loss_val, C

    updates, opt_state = optimizer.update(grads, opt_state)
    model = eqx.apply_updates(model, updates)

    return model, opt_state, loss_val, C

def train_model(
    model,
    coords,
    areas,
    target_density,
    penalty,
    opt_state=None,
    num_epochs=100,
    lr=1e-4,
    patience=2,
    min_delta=0.005,
):
    """
    Train a single model with augmented Lagrangian volume constraint
    and early stopping based on the loss.
    """

    optimizer = optax.chain(
        optax.clip_by_global_norm(1.0),
        optax.adabelief(lr),
    )
    if opt_state is None:
        opt_state = optimizer.init(eqx.filter(model, eqx.is_array))

    lam = 300.0 

    best_loss = np.inf
    epochs_no_improve = 0

    for epoch in tqdm(range(num_epochs), desc="Epochs"):

        model, opt_state, loss_val, C = optimisation_step(
            model,
            opt_state,
            optimizer,
            coords,
            areas,
            target_density,
            lam,
            penalty,
        )

        # Update lambda if constraint is finite
        if np.isfinite(C):
            lam = lam + 2.0 * penalty * C

        loss_scalar = float(loss_val)
        print(
            f"Epoch {epoch:03d} | loss = {loss_scalar:.6f} | C = {float(C):.6f} | lam = {float(lam):.6f}"
        )

        # ----- Early stopping logic -----
        if best_loss - loss_scalar > min_delta:
            best_loss = loss_scalar
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print(
                    f"Early stopping at epoch {epoch:03d} "
                    f"(no significant improvement for {patience} epochs; "
                    f"best loss = {best_loss:.6f})"
                )
                # break
        # -------------------------------

    return model, opt_state
```
